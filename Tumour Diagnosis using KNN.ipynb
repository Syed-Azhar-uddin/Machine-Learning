Tumour Diagnosis Using KNN 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
Tumour = pd.read_csv('Tumour.csv')
```

```python
Tumour_Updated=pd.get_dummies(Tumour,{'diagnosis'})
```

```python
Tumour_Updated.head()
```

```python
X=Tumour_Updated.drop(['id','Unnamed: 32','diagnosis_B','diagnosis_M'],axis=1)
```

```python
y=Tumour_Updated['diagnosis_B']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler =StandardScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model = KNeighborsClassifier(n_neighbors=1)
```

```python
model.fit(X_train,y_train)
```

```python
pred =model.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,pred))
```

```python
print(confusion_matrix(y_test,pred))
```

```python
error_rate = []

for i in range(1,60):
    
    model = KNeighborsClassifier(n_neighbors=i)
    model.fit(X_train,y_train)
    pred_i =model.predict(X_test)
    error_rate.append(np.mean(pred_i != y_test))
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,60),error_rate,color='blue',marker='o',markersize=10)
plt.title("Error rate vs K")
plt.xlabel('K')
plt.ylabel('Error rate')
```

```python
model = KNeighborsClassifier(n_neighbors=17)
model.fit(X_train,y_train)
pred =model.predict(X_test)
print(classification_report(y_test,pred))
```
